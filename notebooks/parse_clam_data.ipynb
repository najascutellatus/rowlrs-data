```python
import pandas as pd
from pathlib import Path
```

```python
cwd = Path.cwd()
cwd
```

```python
data_path = cwd.parent / 'data' / 'clam' 
data_path
```

```python
cfile = data_path / "ExampleClamData.xlsx"
cfile
```

```python
# # %%
# trip_dtype = {
#     "Start_Lat": float,
#     "Start_Lon": float,
#     "Stop_Lat": float,
#     "Stop_Lon": float,
#     "ByCatch?": bool,
# }

# with open(cfile) as f:
data_trip = pd.read_excel(cfile, sheet_name="TripLevelData", parse_dates={"time": [0, 1, 2, 3]})
data_trip
```

```python
data_tow = pd.read_excel(cfile, sheet_name="TowLevelData")
data_tow.head()
```

```python
merged = pd.merge(data_trip, data_tow, on="TowID")
merged.head()
# data_trip.merge(data_tow, how="cross", on="TowID")
```

```python
merged
```

```python
merged.reset_index(inplace=True)
merged.set_index(['time'], inplace=True)
merged
```

```python
ds = merged.to_xarray()
ds
```
