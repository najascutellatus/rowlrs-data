```python
import pandas as pd
from pathlib import Path
```

```python
cwd = Path.cwd()
cwd
```

```python
file_path = cwd.parent / 'data' / 'eDNA' / 'profiles_sound' / 'example_ysi_vertical_profile_20190814.xlsx'
file_path
```

```python
df = pd.read_excel(file_path, 0, parse_dates=[['Date', 'Time']])
df.head()
```

```python
rename_dict = {}
rename_dict["Date_Time"] =  "time"
rename_dict["Site"] =  "site"
rename_dict["Unit ID"] =  "unit_id"
rename_dict["User ID"] =  "user_id"
rename_dict["°C"] =  "temperature" #sea_water_temperature (K)
rename_dict["mmHg"] =  "pressure" #sea_water_pressure (dbar)
rename_dict["DO %"] =  "dissolved_oxygen_percentage" # mass_concentration_of_oxygen_in_sea_water (kg m-3)
rename_dict["DO mg/L"] =  "dissolved_oxygen_concentration" # mass_concentration_of_oxygen_in_sea_water (kg m-3)
rename_dict["SPC-uS/cm"] =  "SPC"
rename_dict["C-mS/cm"] =  "conductivity" #sea_water_electrical_conductivity
rename_dict["TDS g/L"] =  "total_dissolved_solids"
rename_dict["SAL-ppt"] =  "salinity" # sea_water_salinity (1e-3 or ppt)
rename_dict["pH"] =  "ph" # sea_water_ph_reported_on_total_scale
rename_dict["DEP m"] =  "depth" # depth (m)
```

```python
df = df.rename(columns=rename_dict).set_index(['time', 'depth', 'site'])
df.head()
```

```python
import xarray as xr
```

```python
ds = df.to_xarray()
ds
```

```python
# Unit ID
# ds['unit_id']

# User ID
# ds['user_id']

# Depth
ds['depth'].attrs['standard_name'] = 'depth'
ds['depth'].attrs['units'] = 'm'

# Pressure
ds['pressure'].attrs['standard_name'] = 'sea_water_pressure'
ds['pressure'].attrs['units'] = 'mmHg' # dbar is standard according to cfconventions.org

# Temperature
ds['temperature'].attrs['standard_name'] = 'sea_water_ph_report_on_total_scale'
ds['temperature'].attrs['units'] = '°C'

# Salinity
ds['salinity'].attrs['standard_name'] = 'sea_water_salinity'
ds['salinity'].attrs['units'] = 'ppt'

# Conductivity
ds['conductivity'].attrs['standard_name'] = 'sea_water_electrical_conductivity'
ds['conductivity'].attrs['units'] = 'mS/cm'

# DO - Percentage
ds['dissolved_oxygen_percentage'].attrs['standard_name'] = 'mass_concentration_of_oxygen_in_sea_water_percentage'
ds['dissolved_oxygen_percentage'].attrs['units'] = '%' # dbar is standard according to cfconventions.org

# DO - Concentration
ds['dissolved_oxygen_concentration'].attrs['standard_name'] = 'mass_concentration_of_oxygen_in_sea_water'
ds['dissolved_oxygen_concentration'].attrs['units'] = 'mg/L' # dbar is standard according to cfconventions.org

# pH - Log Scale
ds['ph'].attrs['standard_name'] = 'sea_water_ph_report_on_total_scale'
ds['ph'].attrs['units'] = '1'

# Total Dissolved Solids
ds['total_dissolved_solids'].attrs['units'] = 'g/L'

# SPC-uS/cm
ds['SPC'].attrs['units'] = 'uS/cm'
```

```python
ds
```

```python
ds.to_netcdf(f"{file_path}.nc")
```
