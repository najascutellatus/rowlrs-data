```python
import pandas as pd
from pathlib import Path
```

```python
cwd = Path.cwd()
cwd
```

```python
data_path = cwd.parent / 'data' / 'fieldwork_log' / 'Fieldwork_Log.xlsx'
data_path
```

```python
df_field_log = pd.read_excel(data_path, sheet_name="Fieldlog", usecols='A:S').drop(['Participating Personnel'], axis=1)
df_field_log.head()
```

```python
df_field_log.columns = ['_'.join(key.lower().split()) for key in df_field_log.keys()]
```

```python
import datetime as dt
```

```python
df_field_log['time_out'] = df_field_log['time_out'].fillna(dt.time(0, 0, 0))
df_field_log['time_in'] = df_field_log['time_in'].fillna(dt.time(0, 0, 0))
df_field_log['datetime_out'] = pd.to_datetime(df_field_log['date_out'].apply(str) + ' ' + df_field_log['time_out'].apply(str))
df_field_log['datetime_in'] = pd.to_datetime(df_field_log['date_in'].apply(str) + ' ' + df_field_log['time_in'].apply(str))
df_field_log.head()
```

```python
df_field_log.set_index(['season'])
```

```python
df_glider_log = pd.read_excel(data_path, sheet_name="Glider")
df_glider_log
```

```python
pd.merge(df_field_log, df_glider_log, left_on="Date Out", right_on="deployment date")
```
